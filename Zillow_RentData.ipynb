```python
#importing the list of ZIP codes
import pandas as pd

url1 = 'https://raw.githubusercontent.com/databootcampNYUss18/REproject/master/'
url2 = 'zipcode_list.csv'
url = url1 + url2

zipList = pd.read_csv(url)
zipList.head(3)
```

```python
#importing the Zillow data for single family and multifamily rentals
#Zillow Rent Index (ZRI): A smoothed, seasonally adjusted measure of the median estimated market rate rent 
#across a given region and housing type. ZRI is a dollar-denominated alternative to repeat-rent indices.

url3 = 'Zip_Zri_AllHomesPlusMultifamily.csv'

ZillowUrl = url1 + url3

Zillow = pd.read_csv(ZillowUrl)

#40 rows (aka ZIP codes included), 95 columns
NewYorkCounty = Zillow[Zillow['CountyName']=='New York']

#58 rows (aka ZIP codes included), 95 columns
QueensCounty = Zillow[Zillow['CountyName']=='Queens']

#41 rows (aka ZIP codes included), 95 columns
BrooklynCounty = Zillow[Zillow['CountyName']=='Kings']    #did you know - Brooklyn is in Kings County

```

```python
rentdata = pd.concat([NewYorkCounty, QueensCounty, BrooklynCounty])
```

```python
rentdata
```

```python
# We analyze the rentdata for the last 5 years: 2013-02 to 2018-02. 
rentdata1 = rentdata[['RegionName','City','State','Metro','CountyName',
      '2013-02','2013-03','2013-04','2013-05','2013-06','2013-07','2013-08','2013-09','2013-10','2013-11','2013-12',
      '2014-01','2014-02','2014-03','2014-04','2014-05','2014-06','2014-07','2014-08','2014-09','2014-10','2014-11','2014-12',
      '2015-01','2015-02','2015-03','2015-04','2015-05','2015-06','2015-07','2015-08','2015-09','2015-10','2015-11','2015-12',
      '2016-01','2016-02','2016-03','2016-04','2016-05','2016-06','2016-07','2016-08','2016-09','2016-10','2016-11','2016-12',
      '2017-01','2017-02','2017-03','2017-04','2017-05','2017-06','2017-07','2017-08','2017-09','2017-10','2017-11','2017-12',
      '2018-01','2018-02'
]]

rentdata1
```

```python
# Set RegionName as index 
rentdata2 = rentdata1.set_index('RegionName')
rentdata2
```

```python
# The compound annual growth rate (CAGR) is the mean annual growth rate of an investment over a specified period of time longer than one year.
# ((ending home value/starting home value)**(1/5))-1

CAGR_RD_home=[]
for index, row in rentdata2.iterrows():
    cagr_rentvalue=(row['2018-02']/row['2013-02'])**(1/5)-1
    CAGR_RD_home.append(cagr_rentvalue)
```

```python
rentdata2['CAGR_RD']=CAGR_RD_home
rentdata2.sort_values('CAGR_RD',ascending=False).head(10)
```

```python
topRD = (rentdata2.sort_values('CAGR_RD',ascending=False).head(10))
topRD
```

```python
# Plot the Top 10 Zip Codes by CAGR for Home Rentals in horizontal bar graph. 
# Zip 11213 (Crown Heights) in Brooklyn has the fastest growing Home Rental rate a.k.a CAGR value

topRD['CAGR_RD'].sort_values().plot(kind='barh',figsize=(5,5),title='CAGR for Rent Data By Zip Code')
```
