```python
#importing the list of ZIP codes
import pandas as pd
import matplotlib.pyplot as plt        # graphics module  


%matplotlib inline  

url1 = 'https://raw.githubusercontent.com/databootcampNYUss18/REproject/master/'
url2 = 'zipcode_list.csv'
url = url1 + url2

zipList = pd.read_csv(url)
zipList.head(3)
```

```python
# Importing Zillow Home Value Index (ZHVI) CSV file. 
# Each Zillow Home Value Index (ZHVI) is a time series tracking the monthly median home value in a particular geographical region. 
# The ZHVI is generated at the zip code level. 
burl1 = 'https://raw.githubusercontent.com/databootcampNYUss18/REproject/master/'
burl2 = 'Zip_Zhvi_AllHomes%20(1).csv'
burl = burl1 + burl2
zhvi = pd.read_csv(burl)
zhvi.head(3)
```

```python
# We analyze the ZHVI for the last 5 years: 2013-02 to 2018-02. 
zhvi1=zhvi[['RegionName','City','State','Metro','CountyName',
      '2013-02','2013-03','2013-04','2013-05','2013-06','2013-07','2013-08','2013-09','2013-10','2013-11','2013-12',
      '2014-01','2014-02','2014-03','2014-04','2014-05','2014-06','2014-07','2014-08','2014-09','2014-10','2014-11','2014-12',
      '2015-01','2015-02','2015-03','2015-04','2015-05','2015-06','2015-07','2015-08','2015-09','2015-10','2015-11','2015-12',
      '2016-01','2016-02','2016-03','2016-04','2016-05','2016-06','2016-07','2016-08','2016-09','2016-10','2016-11','2016-12',
      '2017-01','2017-02','2017-03','2017-04','2017-05','2017-06','2017-07','2017-08','2017-09','2017-10','2017-11','2017-12',
      '2018-01','2018-02'
]]

zhvi1
```

```python
# Select zip codes in New York City. 
# RegionName column is zip code. 
zhvi2 = zhvi1[(zhvi1['City'] =='New York') & (zhvi1['State'] =='NY')& (zhvi1['Metro']=='New York')]
zhvi2
```

```python
# Select zip codes in counties of New York, Kings, and Queens. 
zhvi3=zhvi2[(zhvi2['CountyName']=='New York')  | (zhvi2['CountyName']=='Kings')  | (zhvi2['CountyName']=='Queens')]
zhvi3
```

```python
# Set RegionName as index 
zhvi4 = zhvi3.set_index('RegionName')
zhvi4
```

```python
# Check data type of each column 
zhvi4.dtypes
```

```python
# Find the average value for each zip code by row. 
zhvi4['AVG_ZHVI']=zhvi4.mean(axis=1)
zhvi4
```

```python
# Plot the average home value index by zip code in horizontal bar graph. 
# The x-axis is the ZHVI in dollars. 
zhvi4['AVG_ZHVI'].sort_values().plot(kind='barh',figsize=(20,20),title='Average Home Value Index By Zip Code')
```

```python
# The top 10 zip codes with the highest average ZHVI are all in New York county. 
# The zip code with the highest average ZHVI is 10013. 
zhvi4[['AVG_ZHVI','CountyName',]].sort_values('AVG_ZHVI').tail(10)
```
